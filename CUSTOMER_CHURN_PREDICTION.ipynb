```python
# Dropping unnecessary columns 'RowNumber', 'CustomerId', and 'Surname' from the DataFrame 'data'
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)

# Displaying the updated DataFrame 'data'
data
```

**Changing categorical values to numerical values**

```python
import pandas as pd

# Creating dummy variables for categorical features and dropping the first category to avoid multicollinearity
data = pd.get_dummies(data, drop_first=True)

# Displaying the first few rows of the updated DataFrame
data.head()

# Converting the DataFrame to integer type for consistency
data = data.astype(int)

# Displaying the updated DataFrame with all values converted to integer type
data
```

**Data Visualization**

```python
# Counting the occurrences of each unique value in the 'Exited' column
exit_counts = data['Exited'].value_counts()

# Displaying the counts of unique values in the 'Exited' column
exit_counts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the figure size
plt.figure(figsize=(8, 6))

# Creating a count plot for the 'Exited' column
sns.countplot(x='Exited', data=data)

# Displaying the plot
plt.show()
```

```python
# Separating the features (X) and the target variable (y)
X = data.drop('Exited', axis=1)
y = data['Exited']
```

```python
# Libraries for data preprocessing and splitting
from sklearn.model_selection import train_test_split

# Library for data scaling
from sklearn.preprocessing import StandardScaler

# Libraries for different machine learning algorithms
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsClassifier

# Libraries for evaluation metrics
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

# Library for encoding categorical variables
from sklearn.preprocessing import LabelEncoder
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Displaying the shapes of the training and testing sets
print('Training Shape: ', X_train.shape)
print('Testing Shape: ', X_test.shape)
```

```python
# Initializing the StandardScaler
scaler = StandardScaler()

# Scaling the training features
X_train_scaled = scaler.fit_transform(X_train)

# Scaling the testing features
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

**Logistic Regression**

```python
# Defining the threshold for classification
threshold = 0.5

# Classifying the training labels based on the threshold
y_train_classified = [1 if value > threshold else 0 for value in y_train]

# Initializing the Logistic Regression model
LR = LogisticRegression()

# Fitting the Logistic Regression model to the scaled training data and classified labels
LR.fit(X_train_scaled, y_train_classified)
```

```python
# Classifying the testing labels based on the threshold
y_test_classified = [1 if value > threshold else 0 for value in y_test]

# Calculating the accuracy of the Logistic Regression model on the scaled testing data
accuracy1 = LR.score(X_test_scaled, y_test_classified)

# Displaying the accuracy of the Logistic Regression model
print("Model Accuracy:", accuracy1)
```

```python
# Importing the Support Vector Machine (SVM) classifier from scikit-learn
from sklearn import svm

# Defining the threshold for classification
threshold = 0.5

# Classifying the training labels based on the threshold
y_train_classified = [1 if value > threshold else 0 for value in y_train]

# Initializing the Support Vector Machine (SVM) classifier
svm_classifier = svm.SVC()

# Fitting the SVM classifier to the scaled training data and classified labels
svm_classifier.fit(X_train_scaled, y_train_classified)
```

```python
# Classifying the testing labels based on the threshold
y_test_classified = [1 if value > threshold else 0 for value in y_test]

# Calculating the accuracy of the SVM classifier on the scaled testing data
accuracy2 = svm_classifier.score(X_test_scaled, y_test_classified)

# Displaying the accuracy of the SVM classifier
print("Model Accuracy:", accuracy2)
```
